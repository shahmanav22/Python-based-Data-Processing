```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
print("Shape:", df.shape)

print("\nColumns:\n")
print(df.columns.tolist())

print("\nData Types:\n")
print(df.dtypes)
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

df["TotalCharges"].dtype
```

```python
df.isnull().sum()
```

```python
df["TotalCharges"].fillna(df["TotalCharges"].median(), inplace=True)

df.isnull().sum()
```

```python
customer_info = df[[
    "gender",
    "SeniorCitizen",
    "tenure",
    "MonthlyCharges",
    "TotalCharges",
    "Contract",
    "PaymentMethod",
    "Churn"
]]

customer_info.head()
```

```python
churned = df[df["Churn"] == "Yes"]

churned.head()

print("Total Churned Customers:", len(churned))
```

```python
high_bill = df[df["MonthlyCharges"] > 80]

high_bill.head()
```

```python
senior = df[df["SeniorCitizen"] == 1]

senior.head()
```

```python
df.sort_values(by="MonthlyCharges", ascending=False).head(10)
```

```python
df.sort_values(by="tenure").head(10)
```

```python
df["YearlyCharges"] = df["MonthlyCharges"] * 12

df[["MonthlyCharges", "YearlyCharges"]].head()
```

```python
def category(tenure):

    if tenure < 12:
        return "New"

    elif tenure < 36:
        return "Regular"

    else:
        return "Loyal"


df["CustomerCategory"] = df["tenure"].apply(category)

df.head()
```

```python
df["Contract"].unique()
```

```python
df["PaymentMethod"].nunique()
```

```python
df["PaymentMethod"].value_counts()
```

```python
contract = df.groupby("Contract")["MonthlyCharges"].mean()

contract
```

```python
df.groupby("Churn")[["MonthlyCharges", "TotalCharges"]].mean()
```

```python
df.groupby("InternetService")["MonthlyCharges"].mean()
```

```python
pivot = pd.pivot_table(
    df,
    values="MonthlyCharges",
    index="Contract",
    columns="Churn",
    aggfunc="mean"
)

pivot
```

```python
df.rename(columns={
    "MonthlyCharges":"Monthly_Charges",
    "TotalCharges":"Total_Charges"
}, inplace=True)

df.head()
```

```python
df["Discount"] = df["Monthly_Charges"] * 0.10

df.head()
```

```python
top = df.nlargest(10, "Monthly_Charges")

top
```

```python
bottom = df.nsmallest(10, "Monthly_Charges")

bottom
```

```python
df.to_csv("Processed_Telco_Data.csv", index=False)

print("Processed dataset saved successfully!")
```
